### Tratamento de features:

#### Carregando Dependências:

```python
import matplotlib.pyplot as plt
import numpy as np
import os
import pickle
```

#### Temperatura Crítica:

```python
fig, axes = plt.subplots(2, 2, figsize=(16, 8))
axes = axes.flatten()

feature = "Temperatura Crítica"

# Formato array:
crit_temp_array = np.array(train_df[lista_features[0]])
plot_original(crit_temp_array, feature)

# Tratamento
crit_temp_array = np.sqrt(crit_temp_array)
plot_tratado(crit_temp_array, feature)

plt.tight_layout()
plt.show()

# Garante que a pasta existe
os.makedirs("features_tratadas", exist_ok=True)

# Salva a variável transformada
with open('features_tratadas/critic_temp.pkl', 'wb') as f:
    pickle.dump(crit_temp_array, f)
```
